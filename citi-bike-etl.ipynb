```python
import pandas as pd
import numpy as np
import seaborn as sns
import glob
from datetime import datetime
```

```python
def parseDate(date):
    for fmt in ('%m/%d/%Y', '%m/%d/%y'):
        try:
            return datetime.strptime(date, fmt)
        except ValueError:
            pass
    raise ValueError('no valid date format found')
```

```python
path ="/Users/chuamelia/Google Drive/Forecasting Time Series/citi-bike/raw-data/*.csv"
all_csvs = glob.glob(path)
```

```python
bikeDF = pd.DataFrame()
list_ = []
for csv in all_csvs:
    df = pd.read_csv(csv,index_col=None, header=0)
    list_.append(df)
bikeDF = pd.concat(list_)
```

```python
columns = {'Date':'Date',
 '24-Hour Passes Purchased (midnight to 11:59 pm)':'24HPass',
 '3-Day Passes Purchased (midnight to 11:59 pm)':'3DPass',
 '7-Day Passes Purchased (midnight to 11:59 pm)':'7DPass',
 'Trips over the past 24-hours (midnight to 11:59pm)':'Trips',
 'Miles traveled today (midnight to 11:59 pm)':'Miles'}
shorten_columns = list(columns.keys())
```

```python
bikeDF 
```

```python
bikeDF['Date'] = bikeDF['Date'].apply(parseDate)
```

```python
newBikeDF = bikeDF[shorten_columns].sort_values(['Date'])
newBikeDF = newBikeDF.rename(columns=columns)
newBikeDF['Year'] = newBikeDF['Date'].map(lambda x: x.year)
newBikeDF['Month'] = newBikeDF['Date'].map(lambda x: x.month)
newBikeDF['Day'] = newBikeDF['Date'].map(lambda x: x.day)
newBikeDF['MonthYear'] = newBikeDF['Date'].map(lambda x: x.strftime('%Y-%m-01'))
newBikeDF.head(5)
```

```python
aggTripsMonthly = pd.DataFrame(newBikeDF.groupby(['Year', 'Month', 'MonthYear']).agg({'Trips': np.mean,'Miles': np.mean,'24HPass' })).reset_index()
aggTripsMonthly.head(5)
```

```python
aggTripsMonthlyLookup = pd.DataFrame((newBikeDF.groupby(['Month']).agg({'Trips': np.mean}))).reset_index()
aggTripsMonthlyLookup.head(5)
```

```python
def removeSeasonalComponent(Month , df):
    mean = aggTripsMonthlyLookup[aggTripsMonthlyLookup['Month'] == Month]['Trips'].iloc[0]
    aggTripsMonthly.loc[aggTripsMonthly['Month'] == Month, 'woSeasonal'] = aggTripsMonthly['Trips'] - mean
```

```python
aggTripsMonthly['woSeasonal'] = 0.0

for Month in range(1,13):
    removeSeasonalComponent(Month , aggTripsMonthly)

aggTripsMonthly['woSeasonal'] = aggTripsMonthly['woSeasonal'] - (aggTripsMonthly['woSeasonal'].min()-1)

aggTripsMonthly.head(5)
```
